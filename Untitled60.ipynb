```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
signals = pd.read_csv("signal_matrix.csv")
ratings = pd.read_csv("human_ratings.csv")

# Add output_id to signals
signals["output_id"] = range(1, len(signals) + 1)

# Merge
df = signals.merge(ratings, on="output_id")

# Quick correlation
print("Correlation with human rating:")
print(df[["mean_token_entropy", "self_perplexity", "rating"]].corr()["rating"])

# Plot
plt.scatter(df["self_perplexity"], df["rating"])
plt.xlabel("Self Perplexity")
plt.ylabel("Human Originality Rating")
plt.title("Perplexity vs Human-Rated Originality (Pilot)")
plt.savefig("pilot_correlation.png")
plt.show()
```
